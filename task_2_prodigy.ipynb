```python
# Import necessary libraries
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Load the train dataset
train_df = pd.read_csv("/content/drive/MyDrive/Prodigy /TASK 2/titanic (1)/train.csv")
```

```python
# Display the first few rows of the dataset
print(train_df.head())
```

**Perform data cleaning**

```python
# Check for missing values
print(train_df.isnull().sum())
```

```python
import numpy as np

# Replace blank values with NaN
train_df.replace('', np.nan, inplace=True)


```

```python
# Impute missing values in the 'Age' column with the mean age
mean_age = train_df['Age'].mean()
train_df['Age'].fillna(mean_age, inplace=True)

# Impute missing values in the 'Cabin' column with the Unknown
train_df['Cabin'].fillna('Unknown', inplace=True)

# Impute missing values in the 'Embarked       ' column with the Unknown
train_df['Embarked'].fillna('Unknown', inplace=True)
```

```python
# Confirm if blank values have been replaced with NaN
print(train_df.isnull().sum())
```

```python
print(pd.DataFrame(train_df.head(10)))
```

```python
# Summary statistics
print(train_df.describe())
```

**exploratory data analysis (EDA)**

Survival Distribution

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Set the style
sns.set_style("whitegrid")

# countplot
ax = sns.countplot(x='Survived',  data=train_df, palette=['lightcoral', 'lightgreen'])

# Add annotations
total = len(train_df)
for p in ax.patches:
    height = p.get_height()
    ax.text(p.get_x() + p.get_width() / 2., height + 5, '{:.0f}'.format(height), ha="center")

# Set title and labels
plt.title('Survival Distribution')
plt.xlabel('Survived')
plt.ylabel('Count of Passengers')

# Set legend
ax.legend(['Non Survived', 'Survived'])

# Show plot
plt.show()
```

- There are three bars addressing traveler classes named as "1", "2", and "3".

- Each bar is marked with both count and level of absolute travelers:  - Class "1" bar is more limited, shaded light blue, showing a count of "216" travelers which is "24.2%" of aggregate.
 - Class "2" bar is much more limited than class "1", hued light blue too, showing a count of "184" which makes up "20.7%" of aggregate.
 - Class "3" bar is essentially taller than others, dull blue tone shows a count of "491", making up "55.1%"of complete travelers.
  - The y-pivot addresses 'Traveler Count' going from '0' to '500'.
   - The x-pivot addresses 'Traveler Class'. - This proposes that most of the travelers were in class 3, with less in classes 1 and 2.


**Passenger Class Distribution**

```python


# Set the style of seaborn
sns.set_style("whitegrid")

# Plot the countplot
ax = sns.countplot(x='Pclass', data=train_df, palette='Blues')

# Add annotations for count above the bars
total = len(train_df)
for p in ax.patches:
    height = p.get_height()
    ax.text(p.get_x() + p.get_width() / 2., height + 10, '{:.0f}'.format(height), ha="center")

# Add annotations for percentage comparison
total_passengers = len(train_df)
for p in ax.patches:
    height = p.get_height()
    percentage = (height / total_passengers) * 100
    ax.text(p.get_x() + p.get_width() / 2., height + 40, '{:.1f}%'.format(percentage), ha="center")

# Set title and labels
plt.title('Passenger Class Distribution')
plt.xlabel('Passenger class')
plt.ylabel('Passenger Count')

# Show plot
plt.show()
```

**Passenger Gender Distribution**

```python

# Set the style of seaborn
sns.set_style("whitegrid")

# Set figure size
plt.figure(figsize=(8, 6))

# Plot the countplot for Sex distribution
ax1 = sns.countplot(x='Sex', data=train_df, palette='Purples')

# Add annotations for count above the bars (Sex distribution)
total_passengers = len(train_df)
for p in ax1.patches:
    height = p.get_height()
    ax1.text(p.get_x() + p.get_width() / 2., height + 10, '{:.0f}'.format(height), ha="center")

# Add annotations for percentage comparison (Sex distribution)
for p in ax1.patches:
    height = p.get_height()
    percentage = (height / total_passengers) * 100
    ax1.text(p.get_x() + p.get_width() / 2., height + 40, '{:.1f}%'.format(percentage), ha="center")

# Set title and labels for Sex distribution
plt.title('Passenger Gender Distribution')
plt.xlabel('Gender ')
plt.ylabel('Passenger Count')

# Set y-axis limit
plt.ylim(0, 700)

# Show plot for Sex distribution
plt.show()

```

**Survival Rate by Passenger Class**

```python
import seaborn as sns

# Set the style of seaborn
sns.set_style("whitegrid")

# Plot the point plot for Survival Rate by Passenger Class
sns.pointplot(x='Pclass', y='Survived', data=train_df, ci=None, color='skyblue')

# Set title and labels for Survival Rate by Passenger Class
plt.title('Survival Rate by Passenger Class')
plt.xlabel('Passenger Class')
plt.ylabel('Survival Rate')
# Set y-axis limit
plt.ylim(0, 1.00)

# Show plot
plt.show()
```

**Survival Rate by Gender**

```python
import matplotlib.pyplot as plt

# Calculate survival rates by gender
survival_rate_by_gender = train_df.groupby('Sex')['Survived'].mean()

# Create a pie chart
plt.figure(figsize=(6, 6))
plt.pie(survival_rate_by_gender, labels=survival_rate_by_gender.index, autopct='%1.1f%%', colors=['lightcoral', 'lightblue'])
plt.title('Survival Rate by Gender')
plt.axis('equal')

# Show plot
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Set the style of seaborn
sns.set_style("whitegrid")

# Filter data for parents and children
parents_df = train_df[train_df['Parch'] > 0]
children_df = train_df[train_df['Parch'] == 0]

# Plot the point plots for Survival Rate by Number of Parents/Children Aboard
sns.pointplot(x='Parch', y='Survived', data=parents_df, ci=None, color='lightblue', label='Parents')
sns.pointplot(x='Parch', y='Survived', data=children_df, ci=None, color='lightcoral', label='Children')

# Set title and labels for Survival Rate by Number of Parents/Children Aboard
plt.title('Survival Rate by Number of Parents/Children Aboard')
plt.xlabel('Number of Parents/Children Aboard')
plt.ylabel('Survival Rate')

# Show legend
plt.legend()

# Show plot
plt.show()

```

- The chart analyzes the endurance paces of guardians and youngsters in light of the quantity of relatives on board.
- The x-hub addresses the "Quantity of Guardians/Kids On board" going from 0 to 6.
- The y-hub addresses the "Endurance Rate" going from 0.0 to 0.6.
- The guardians' endurance rate increments with 1 to 3 relatives on board, then forcefully declines at 4.
- The kids' line has just a single piece of information at '0', demonstrating a particular endurance rate.

# trends and pattern

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Set the style
sns.set_style("whitegrid")

# Create a figure and subplots
fig, axes = plt.subplots(2, 1, figsize=(6, 8))

# Plot 1: Survival Distribution
ax1 = sns.countplot(x='Survived',  data=train_df, palette=['lightcoral', 'lightgreen'], ax=axes[0])

# Add annotations for count above the bars (Survival Distribution)
total = len(train_df)
for p in ax1.patches:
    height = p.get_height()
    ax1.text(p.get_x() + p.get_width() / 2., height + 5, '{:.0f}'.format(height), ha="center")

# Set title and labels for Survival Distribution
ax1.set_title('Survival Distribution')
ax1.set_xlabel('Survived')
ax1.set_ylabel('Count of Passengers')

# Set legend for Survival Distribution
ax1.legend(['Non Survived', 'Survived'])

# Plot 2: Passenger Gender Distribution
ax2 = sns.countplot(x='Sex', data=train_df, palette='Purples', ax=axes[1])

# Add annotations for count above the bars (Passenger Gender Distribution)
total_passengers = len(train_df)
for p in ax2.patches:
    height = p.get_height()
    ax2.text(p.get_x() + p.get_width() / 2., height + 10, '{:.0f}'.format(height), ha="center")

# Add annotations for percentage comparison (Passenger Gender Distribution)
for p in ax2.patches:
    height = p.get_height()
    percentage = (height / total_passengers) * 100
    ax2.text(p.get_x() + p.get_width() / 2., height + 40, '{:.1f}%'.format(percentage), ha="center")

# Set title and labels for Passenger Gender Distribution
ax2.set_title('Passenger Gender Distribution')
ax2.set_xlabel('Gender ')
ax2.set_ylabel('Passenger Count')

# Set y-axis limit for Passenger Gender Distribution
ax2.set_ylim(0, 700)

# Show plot
plt.tight_layout()
plt.show()
```

Here are the central issues contrasting the two charts: - **Endurance Dissemination versus Orientation Distribution**:
  - **Endurance Distribution**: More travelers didn't get by (549) than the individuals who made due (342). This is addressed by a red bar for non-survivors and a green bar for survivors.
  - **Orientation Distribution**: There were altogether more male travelers (64.8% or 577 people) than female travelers (35.2% or 314 people). This is addressed by a blue bar for guys and a purple bar for females.
  - **Comparison**: While the endurance diagram shows a bigger number of non-survivors, the orientation chart shows a larger number of guys.
  * Possible Relationship: an expected connection among orientation and endurance rate. For instance, on the off chance that a critical part of the survivors were female, this could recommend that ladies were bound to make due than men. This could be because of elements,
  for example, ladies and kids being given need for life-saving gear or being situated in pieces of the boat that were more secure.

```python
# Set up the figure and axes
plt.figure(figsize=(10, 6))

# Plot the KDE plot for Age distribution
sns.kdeplot(train_df['Age'].dropna(), shade=True, color='skyblue')

# Set title and labels for Age distribution
plt.title('Age Distribution (Kernel Density Estimation)')
plt.xlabel('Age')
plt.ylabel('Density')

# Show plot for Age distribution
plt.show()
```

Here are  the trend demonstrated by the graph:

- The graph is a **Kernel Density Estimation** of age distribution.
- There is a **peak in density** around the age of 40.
- The density **decreases** as the age increases past 40 and also is lower before 40.
- This suggests that the majority of the individuals are around the age of 40, with fewer individuals in the younger and older age groups.


```python
import seaborn as sns
import matplotlib.pyplot as plt

# Set the style of seaborn
sns.set_style("whitegrid")

# Method 1: Stacked Bar Plot
plt.figure(figsize=(10, 5))
sns.countplot(x='SibSp', hue='Survived', data=train_df, palette=['#FF6347', '#7CFC00'])
plt.title('Survival Count by Number of Siblings/Spouses Aboard')
plt.xlabel('Number of Siblings/Spouses Aboard')
plt.ylabel('Count')
plt.legend(title='Survived', labels=['No', 'Yes'])
plt.show()

```

 the trend demonstrated by the graph:

- The graph is a **bar chart** titled "Survival Count by Number of Siblings/Spouses Aboard".
- There are two sets of bars, **green** and **red**, representing those who survived (Yes) and those who did not survive (No), respectively.
- Most people, especially those without siblings/spouses aboard, did not survive as indicated by the prominent red bar at '0' on the x-axis.
- There's a significant **drop** in both survival and non-survival counts as the number of siblings/spouses increases.
- This suggests that individuals with fewer siblings/spouses aboard had a higher chance of survival. However, individuals with one or more siblings/spouses aboard had a lower survival count.
